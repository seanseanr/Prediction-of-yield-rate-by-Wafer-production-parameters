```python
#mask = (df3['6'] > 100) & (df3['8'] > 100) & (df3['11'] > 100) & (df3['12'] > 100) & (df3['15'] < -10)
def df_exclude_outlier(df_in,*col_names):
    masks = []
    for col_name in col_names:
        q1 = df_in[col_name].quantile(0.25)
        q3 = df_in[col_name].quantile(0.75)
        iqr = q3-q1
        box_low  = q1-1.5*iqr
        box_high  = q3+1.5*iqr
        masks.append((df_in[col_name] > box_low) & (df_in[col_name] < box_high))
    df_out = df_in
    for mask in masks:
        df_out =df_out[mask]
    return df_out   
```
