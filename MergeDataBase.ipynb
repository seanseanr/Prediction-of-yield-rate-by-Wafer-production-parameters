```python
i = 0
#df = pd.DataFrame(columns = ["Lot-Id","Wafer-No","No","6","7","8","11","12","15","ANN"])
#df.columns = ["Lot-Id","Wafer-No","No","6","7","8","11","12","15","ANN"]
df = pd.DataFrame()
for f in ext6_files:
    df6 = pd.read_csv(f, sep="\t", header= None)
    df7 = pd.read_csv(ext7_files[i], sep="\t", header= None)
    df8 = pd.read_csv(ext8_files[i], sep="\t", header= None)
    df11 = pd.read_csv(ext11_files[i], sep="\t", header= None)
    df12 = pd.read_csv(ext12_files[i], sep="\t", header= None)
    df15 = pd.read_csv(ext15_files[i], sep="\t", header= None)
    i += 1
    df6.loc[:,2] = df7.loc[:,1]
    df6.loc[:,3] = df8.loc[:,1]
    df6.loc[:,4] = df11.loc[:,1]
    df6.loc[:,5] = df12.loc[:,1]
    df6.loc[:,6] = df15.loc[:,1]
    l1 = f.split(".",1)
    l2 = l1[0].split("_",1)
    df6.insert(loc=0,column="Lot-ID",value=[int(l2[0]) for _ in range(len(df6.loc[:,0]))])
    df6.insert(loc=1,column="Wafer-NO",value=[int(l2[1]) for _ in range(len(df6.loc[:,0]))])
    df6.insert(loc=9,column="ANN",value=[1 for _ in range(len(df6.loc[:,0]))])
    #df = df6
    #df.columns("Lot-Id","Wafer-No","No","6","7","8","11","12","15","ANN"])
    #print(df6)
    frames = [df, df6]
    df = pd.concat(frames)
df.columns = ["Lot-Id","Wafer-No","No","6","7","8","11","12","15","ANN"]
df.to_csv('out.csv',sep=',',header=False)
#df6 = pd.read_csv(ext6_files[1003], sep="\t", header= None)
#df6 = []
#df6 = pd.read_csv(ext6_files[1003], sep="\t", header= None)
```
